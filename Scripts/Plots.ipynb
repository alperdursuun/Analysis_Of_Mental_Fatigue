```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_Neuro=pd.read_csv("Kişi.1\Morning\ScanResult.csv")
df_BVP=pd.read_csv("Kişi.1\Morning\BVP.csv")
df_ACC=pd.read_csv("Kişi.1\Morning\ACC.csv")
df_EDA=pd.read_csv("Kişi.1\Morning\EDA.csv")
df_HR=pd.read_csv("Kişi.1\Morning\HR.csv")
df_IBI=pd.read_csv("Kişi.1\Morning\IBI.csv")
df_TEMP=pd.read_csv("Kişi.1\Morning\TEMP.csv")
```

```python
def plotter(df_input, Hz):
    df_input.plot(kind='line')
    fig = plt.gcf()
    fig.set_size_inches(25, 25)
    measurements = [285*Hz, 300*Hz, 585*Hz, 600*Hz, 885*Hz, 900*Hz, 1185*Hz, 1200*Hz, 1485*Hz, 1500*Hz]
    names = ["Zone1", "Zone1_end", "Zone2", "Zone2_end", "Zone3", "Zone3_end",
             "Zone4", "Zone4_end", "Zone5", "Zone5_end"]
    colors = ["green", "green", "blue", "blue", "orange", "orange", "red", "red", "blue", "blue"]
    for measurement, name, color in zip (measurements, names, colors):
        plt.axvline(x = measurement, linestyle = '--', 
                   linewidth = 2.5, label = '{0} at {1}'.format(name, measurement), c = color)
    plt.legend()
```
