```python
def dataCollection(importFile, exportFile, Hz, df_push):

    subject = ['KiÅŸi.1', 'KiÅŸi.2', 'KiÅŸi.3', 'KiÅŸi.4', 'KiÅŸi.5', 'KiÅŸi.6', 'KiÅŸi.7', 'KiÅŸi.8', 'KiÅŸi.9', 'KiÅŸi.10']
    instance=['Morning', 'Evening']

    for k in range(10):
        for s in range(2):        
            if s == 0:        
                filepath1=os.path.join(subject[k], instance[s], importFile)
                filepath2=os.path.join(subject[k], instance[s], exportFile)
            elif s == 1:
                filepath1=os.path.join(subject[k], instance[s], importFile)
                filepath2=os.path.join(subject[k], instance[s], exportFile)

            df_pull = pd.read_csv(filepath1)

            root = int(285*Hz)
            limit = root + int(30*Hz)
            for i in range(5):
                for j in range(root, limit):
                    entry = df_pull.iloc[[j], :]
                    df_push = df_push.append(entry,ignore_index=True)                
            
                root = root + int(300*Hz)
                limit = root + int(30*Hz)
            
            df_push.dropna(how='all', axis=1, inplace=True)
            df_push.to_csv(filepath2, index=False)
            df_push.drop(df_push.index, inplace=True)
```

```python
dataCollection('ScanResult.csv', 'neuroScan.csv', 1, df_neuroScan)
```
